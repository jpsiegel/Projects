```python
# Cluster with DBSCAN

import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from geopy.distance import great_circle
from shapely.geometry import MultiPoint

coords = df[["latitude","longitude"]]
print(coords)

kms_per_radian = 6371.0088
epsilon = 8 / kms_per_radian
db = DBSCAN(eps=epsilon, min_samples=4, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
cluster_labels = db.labels_
num_clusters = len(set(cluster_labels))
print(cluster_labels)
clusters = pd.Series([coords[cluster_labels == n] for n in range(-1, num_clusters)])
print('Number of clusters: {}'.format(num_clusters))
dict_of_dfs = {i: clusters[i] for i in range(num_clusters)}
print("Sizes:", [len(i) for i in dict_of_dfs.values()])

def get_color():
  for color in ["red", "green", "blue", "yellow", "purple", "pink", "orange"]:
    yield color
color = get_color()

global ax
first_iter = True
for idx, _df in dict_of_dfs.items():
  if first_iter:
    ax = _df.plot(kind="scatter", x="longitude", y="latitude", alpha=1, color="black") # red means data point without cluster
  else:
    _df.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4, ax=ax, color=next(color))
  first_iter = False
plt.show()
```
